## 1. Prime Number Checker
###  This program checks whether a number entered is a prime number or not.
A prime number is a number that is only divisible by 1 and itself, like 2, 3, 5, 7....

Logic i used  to write the code

--> the program takes an integer input from the user.
It handles cases like

    If the number is less than or equal to 1 --> it is not prime.

then  it checks divisibility

Instead of checking all numbers *it only checks from 2 up to the square root of the number.*

    If any number divides it evenly --> the number is not prime.

    If no divisor is found--> the number is prime.

---



````python
def is_prime(number):
    ''''the number take input -- number
            if number <= 1 --- its not aprime number
            else ckeck for sqrt of the number if it devide by any number in th
              list(from 2 to sqrt(number)) its not a prime number
            else its prime  '''
    if number <= 1:
        return False

    # Check divisibility from 2 to sqrt(number)
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False

    return True


# Simple CLI interaction
def main():

  num = int(input("Enter a number to check if it's prime: "))

  if is_prime(num):
      print(f"{num} is a prime number.")
  else:
      print(f"{num} is not a prime number.")
````

````python
for i in range(1,20):
  print(i,is_prime(i))
````

````python
for i in range(4):
  main()

````

## 2. Fibonacci Sequence Generator

### This program generates a Fibonacci sequence up to a number of terms specified by the user. The Fibonacci sequence is a series where each number is the sum of the previous two numbers.
  
  the logic i used to work

  first --> `The program asks  how many terms they want.`

It

```
  # initializes two variables:  a = 0 , and b = 1
```

thenit uses a loop to generate the sequence:
  

```
# Add the current value (a) to the list
  Update values:
  a becomes b
  b becomes a + b
```

The  process repeats until the desired number of terms is reached.

````python
def fibonacci(n):
    sequence = []

    a, b = 0, 1

    for _ in range(n):
        sequence.append(a)
        a, b = b, a + b

    return sequence



def generate_fibonacci():

  terms = int(input("How many Fibonacci terms? "))

  result = fibonacci(terms)

  print("Fibonacci sequence:")
  print(result)
````

````python
generate_fibonacci()
````

## 3. CLI Calculator

### This is a command-line calculator that allows the user to perform basic arithmetic operations:


*   Addition
*   Subtraction
*   Multiplication
*   Division

It runs continuously until the user chooses to exit.

logic i used to write the code


The program displays a menu of operations.
The user selects an option (1–5).
It `checks if the choice is valid`:

    If not → shows an error and asks again.
    It takes two numbers as input.
Based on the selected option:

    Calls the corresponding function
Special handling:

    Division checks if the denominator is zero to avoid errors.
The program runs inside a loop so the user can perform multiple calculations until they choose to exit




````python
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Error: Division by zero!"
    return x / y


def calculator():
    print("\n--- Simple CLI Calculator ---")

    while True:
        print("\nChoose operation:")
        print("1. Add")
        print("2. Subtract")
        print("3. Multiply")
        print("4. Divide")
        print("5. Exit")

        choice = input("Enter choice (1-5): ")

        if choice == '5':
            print("Exiting calculator. Goodbye!")
            break

        if choice not in ['1', '2', '3', '4']:
            print("Invalid choice. Try again.")
            continue

        try:
            num1 = float(input("Enter first number: "))
            num2 = float(input("Enter second number: "))
        except ValueError:
            print("Invalid input. Please enter numbers.")
            continue

        if choice == '1':
            print("Result:", add(num1, num2))
        elif choice == '2':
            print("Result:", subtract(num1, num2))
        elif choice == '3':
            print("Result:", multiply(num1, num2))
        elif choice == '4':
            print("Result:", divide(num1, num2))


````

````python
# Run program
calculator()
````
