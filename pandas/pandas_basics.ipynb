# Step 2: Data Cleaning
Handle missing values:

    Age → median
    Embarked → mode
    Cabin → drop
    Remove duplicates if any

```python
df.isnull().sum()
```

```python
missing_percent = (df.isnull().sum() / len(df)) * 100
print(missing_percent)
```

from this we can understand that  2 columns age , and cabin have missing values
    
    age= 177 missing (19.865320)
    cabin= 687 missing (77.104377)
    Embarked  = 2 missing  0.224467
we have to use difffrent method for both   

age column have around 20 % missing so we can fill it with mean or meadins

but cabine column have missing greater than 70 % its better to drop this column


```python
# Fill missing Age with median
df["Age"].fillna(df["Age"].median(), inplace=True)

```

```python
# Fill missing Embarked with mode
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)

```

```python
# Drop Cabin column
df.drop(columns=["Cabin"], inplace=True)

```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

# Step 3: Data Analysis
Using groupby():

    Survival rate by gender
    Survival rate by class
    Average age per class
    Survival rate by age group


```python
survival_gender = df.groupby("Sex")["Survived"].mean()
print(survival_gender)
```

```python
survival_gender = df.groupby("Sex")["Survived"].describe()
print(survival_gender)
```

```python
survival_class = df.groupby("Pclass")["Survived"].mean()
print(survival_class)
```

```python
avg_age_class = df.groupby("Pclass")["Age"].mean()
print(avg_age_class)
```

```python
# Create age groups
df["AgeGroup"] = pd.cut(df["Age"],
                       bins=[0, 12, 18, 35, 60, 100],
                       labels=["Child", "Teen", "Young Adult", "Adult", "Senior"])

survival_age_group = df.groupby("AgeGroup")["Survived"].mean()
print(survival_age_group)
```

this implies kids in age group of 0-12 have high survival rate

# Step 4: Filtering
    Female passengers who survived
    Children who survived
    Passengers in 1st class with high survival probability


```python
female_survived = df[(df["Sex"] == "female") & (df["Survived"] == 1)]
female_survived.PassengerId.count()
```

```python
children_survived = df[(df["Age"] < 12) & (df["Survived"] == 1)]
children_survived.Age.value_counts()
```

```python
first_class_survived = df[(df["Pclass"] == 1) & (df["Survived"] == 1)]
first_class_survived.PassengerId.count()
```
